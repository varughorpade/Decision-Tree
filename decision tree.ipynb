```python
import pandas as pd
import numpy as np
import seaborn as sn
```

```python
emp_df=pd.read_csv('employee.csv')
```

```python
emp_df.describe()
```

```python
emp_df.head()
```

```python
#removing redundant variables
print(emp_df.EmployeeCount.unique())
print(emp_df.EmployeeNumber.unique())
print(emp_df.Over18.unique())
print(emp_df.StandardHours.unique())

```

```python
for i in emp_df.columns:
    print(emp_df[i].unique())
```

```python
employee=emp_df.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis=1)
```

```python
employee.shape
```

```python
employee.isna().sum()
```

```python
employee.value_count()
```

```python
employee['Attrition'].unique()
```

```python
employee['Attrition'].replace(['Yes','No'],[1,0],inplace=True)
```

```python
employee['Attrition'].unique()
```

```python
employee.info()
```

```python
numerics = ['int64']

employee_num = employee.select_dtypes(include=numerics)
```

```python
employee_num.info()
```

```python
char = ['object']

employee_char = employee.select_dtypes(include=char)
```

```python
employee_char.info()
```

```python
employee_dummies=pd.get_dummies(employee_char)
```

```python
employee_dummies.head()
```

```python
employee_combined=pd.concat([employee_num,employee_dummies],axis=1)
```

```python
employee_combined.head()
```

```python
employee_combined.info()
```

```python
#divide data into tarin and test
from sklearn.model_selection import train_test_split
```

```python
train_x=employee_combined.drop(['Attrition'],axis=1)
```

```python
train_x.shape
```

```python
train_y=employee_combined['Attrition']
```

```python
train_y.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(train_x,train_y,test_size=0.3,random_state=20)
```

```python
#import train classifier model
from sklearn import tree
```

```python
dt=tree.DecisionTreeClassifier()  #create decision tree object
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred=dt.predict(X_test)
```

```python
#computing the model performance matrics
from sklearn.metrics import accuracy_score, precision_score, recall_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print('Accuracy :', accuracy)
print('precision :', precision)
print('recall :', recall)
```
